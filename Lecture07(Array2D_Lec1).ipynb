### Генерация  массивов

#### Генерируем список и превращаем его в массив


```python
l2=list(range(12))
b = np.array(l2,dtype='int16' )
d=b.reshape(4,3)
print(d)
```

```python
b = np.array(range(-6,6,1),dtype='float32')
print(b.size)
print(b)
d=b.reshape((3,4))
print(d)
```

```python
x=5.0
l=[x+i**2  for i in range(10)]
a=np.array(l).reshape((2,5))
print(a)
```

#### Генерируем одномерный целочисленный массив и превращаем его в 2-d массив с элеметами типа float64

```python
d=np.arange(1,20,2)
print(d)
print(d.dtype)
d1=np.array(d,dtype="float64").reshape((2,5))
print(d1)
d1.dtype
```

#### Генерируем linspace()  одномерный вещественный массив и превращаем его в 2-d массив с элеметами типа float64

```python
d=np.linspace(1,20,25)
print(d.size)
print(np.around(d,3))
d1=d.reshape((5,5))
d1.dtype ='float64'
np.around(d1,2)
```

#### Создаём список с помощью генератора случайных действительных чисел

```python
d=np.random.rand(24)*100
print(d)
d1=d.reshape((6,4))
d1.dtype='float64'
print('------------------------------------')
print(np.around(d1,3))
```

##### Фильтрация массива по индексам  элеметов массива 

```python
b = np.array(range(-6,6,1),dtype='float64' )
d=b.reshape(3,4)
print(d)
d[0:2:1,:] #Извлекаем строки
```

```python
b = np.array(range(-6,6,1),dtype='float64' )
d=b.reshape(3,4)
print(d)
d[:,0:2:1] #Извлекаем столбцы
```
