
#### Чтение/запись  данных из файла методами NumPy

```python
filedata = np.genfromtxt('data1.txt', delimiter=',')
filedata = filedata.astype('int32')
print(filedata)
```

#### Запись данных в файл

```python
np.savetxt('test22.txt', filedata, fmt='%1.4e')   # use exponential notation
filedata1 = np.genfromtxt('test22.txt', delimiter=',')
filedata1 = filedata1.astype('float')
print(filedata1)
```

#### Запись и чтение массива в формате .npy

##### Фильтрация массива по индексам  элеметов массива 

```python
a=np.linspace(1.5,8,12)
print(np.around(a,2)) 
np.around(a[0:4:1],2) #Извлекаем элементы a[0],a[1],a[2],a[3]
```

```python
print(np.around(a,2)) 
np.around(a[0:12:2],2) #Извлекаем элементы с чётными индексами
```

```python
print(np.around(a,2)) 
np.around(a[3:12:3],2) #Извлекаем элементы с индексами кратными 3
```
